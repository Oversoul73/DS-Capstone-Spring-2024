```python
import os

root_dir = r"database\archive"

audio_files = []

for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        if file.endswith('.mp3') or file.endswith('.wav'):
            audio_files.append(os.path.join(subdir, file))

print(f"Found {len(audio_files)} audio files.")
```

```python
# Function to rename files
def rename_files(directory):
    for subdir, dirs, files in os.walk(directory):
        for file in files:
            if '-' in file:
                # Construct old file path
                old_file_path = os.path.join(subdir, file)
                # Generate new file name by replacing '-' with '_'
                new_file_name = file.replace('-', '_')
                # Construct new file path
                new_file_path = os.path.join(subdir, new_file_name)
                # Rename the file
                os.rename(old_file_path, new_file_path)
                print(f"Renamed '{old_file_path}' to '{new_file_path}'")


rename_files(root_dir)
```

```python
root_dir = r"database\archive"

subdir_file_counts = {}

for subdir, dirs, files in os.walk(root_dir):
    if subdir == root_dir:
        continue
    file_count = len(files)
    # Store the count in the dictionary with subdir as key
    subdir_file_counts[os.path.basename(subdir)] = file_count

# Print the counts
for subdir, count in subdir_file_counts.items():
    print(f"{subdir}: {count} files")
```
