```python
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer

# Load pre-trained model and tokenizer
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/wav2vec2-large-960h")
model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-960h")
```

```python
import os

default_directory = r"C:\CodeRepo\DS-Capstone-Spring-2024"
os.chdir(default_directory)
current_directory = os.getcwd()

print("Current working directory:", current_directory)
```

```python


import librosa

audio_input, sample_rate = librosa.load("database\sa01.wav", sr=16000)
input_values = tokenizer(audio_input, return_tensors="pt").input_values
```

```python
import torch
with torch.no_grad():
    logits = model(input_values).logits

predicted_ids = torch.argmax(logits, dim=-1)
transcription = tokenizer.decode(predicted_ids[0])
transcription
```

```python
root_dir = r"database\archive"

transcriptions = {}

for subdir, dirs, files in os.walk(root_dir):
    if 'Suprised' in subdir:
        for file in files:
            full_path = os.path.join(subdir, file)
            audio_input, sample_rate = librosa.load(full_path, sr=16000)
            input_values = tokenizer(audio_input, return_tensors="pt").input_values

            with torch.no_grad():
                logits = model(input_values).logits

            predicted_ids = torch.argmax(logits, dim=-1)
            transcription = tokenizer.decode(predicted_ids[0])
            transcriptions[file] = transcription
print(transcriptions)
```

```python
import json

transcriptions_json = json.dumps(transcriptions, indent=4)

print(transcriptions_json)
```
